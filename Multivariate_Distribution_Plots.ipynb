```
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```
tips = sns.load_dataset('tips')

tips.head()
```

```
fig, ax = plt.subplots(figsize=(10.5, 7.2))

sns.scatterplot(x='total_bill', y='tip', data=tips, color='#F15B2A', ax=ax)
```

```
sns.jointplot(x='total_bill', y='tip', data=tips, color='#F15B2A', kind='reg')
```

```
sns.jointplot(x='total_bill', y='tip', data=tips, color='#F15B2A', kind='hex')
```

```
sns.jointplot(x='total_bill', y='tip', data=tips, color='#F15B2A', kind='kde') #density distribution
```

```
from mpl_toolkits.mplot3d import Axes3D
```

```
fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot_trisurf(tips['total_bill'], tips['tip'], tips['size'])
```

```
fig = plt.figure(figsize=(15,9))
ax = fig.add_subplot(111, projection='3d')

ax.bar3d(tips['total_bill'], tips['tip'], np.zeros(len(tips['size'])), 1, 1, tips['size'], color='#F15B2A')
```

```
sns.boxplot(x='time', y='tip', hue='smoker', data=tips)
```

```
fig, ax = plt.subplots(figsize=(10.5,7.2))

sns.violinplot(x='time', y='tip', hue='smoker', data=tips, ax=ax)
```
