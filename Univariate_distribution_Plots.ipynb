```
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```
np.random.seed(0)
mu, sigma = 0,1

data = np.random.normal(mu, sigma, 1000)
```

```
p = sns.distplot(data, color='#F15B2A',
                 hist_kws=dict(edgecolor='#F15B2A', linewidth=2),
                 bins=50,
                 kde=True,
                 kde_kws=dict(shade=True))
```

```
p = sns.kdeplot(data, 
                color='#F15B2A',
                shade=True)
```

```
p = sns.boxplot(data,
                palette=['#F15B2A'])
for e in p.artists:
  r,g,b,a = e.get_facecolor()
  p.set_facecolor((r,g,b, .3))

for i, artist in enumerate(p.artists):
  artist.set_edgecolor('#F15B2A')

for j in range(0,6):
  line = p.lines[j]
  line.set_color('#F15B2A')
  line.set_mfc('#F15B2A')
  line.set_mec('#F15B2A')
```

```
p = sns.violinplot(data,
                   palette=["#F15B2A"])
```

```
#Using MatPlotLib to change element properties

parts = plt.violinplot(data,
                       vert=False,
                       showmeans=True,
                       showmedians=True)

for p in parts['bodies']:
  p.set_facecolor("#F15B2A")
  p.set_edgecolor("#F15B2A")
  p.set_alpha(0.3)
```
